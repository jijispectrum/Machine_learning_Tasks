```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# Create a plot
plt.plot(x, y, label='Line 1')
plt.title('Sample Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Save the plot to a file
plt.savefig('plot.png', dpi=300, bbox_inches='tight')  # Save as a .png file

# Display the plot (optional)
plt.show()
```
