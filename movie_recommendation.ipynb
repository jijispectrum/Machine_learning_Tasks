```python
import pandas as pd
import numpy as np
df_movie=pd.read_csv('tmdb_5000_movies.csv')
df_movie.head()
```

```python
df_credit=pd.read_csv('tmdb_5000_credits.csv')
df_credit.head()
```

```python
df_credit.columns=['id','title','cast','crew']
movie_credit=df_credit.merge(df_movie,on='id')
movie_credit.head()

```

```python
movie_credit.isna().sum()
```

```python
movie_credit.dropna(inplace=True)
```

```python
movie_credit.isna().sum()
```

```python
#filtering 
c=movie_credit['vote_average'].mean()
c
```

```python
#taking 90% voted movie
m=movie_credit['vote_count'].quantile(0.9)
m
```

```python
#making predictions 
top_movies=movie_credit.copy().loc[movie_credit['vote_count']>=m]
top_movies.shape
```

```python
#calculating weight
def weighted_rating(x,m=m,c=c):
    v=x['vote_count']
    r=x['vote_average']
    return (v/(v+m)*r)+(m/(v+m)*c)
```

```python
top_movies['score'] = top_movies.apply(weighted_rating,axis=1)
top_movies.head(3)
```

```python
top_movies.sort_values('score',ascending=False)
top_movies.head()
```

```python
# top 5 movies
top_movies[['title_x','vote_count','vote_average','score']].head(5)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(stop_words='english')
movie_credit['overview']=movie_credit['overview'].fillna('')

```

```python
tfidf_matrix=tfidf.fit_transform(movie_credit['overview'])
print(tfidf_matrix)
```

```python
#finding cosine similarity 
from sklearn.metrics.pairwise import cosine_similarity
movie_similarity=cosine_similarity(tfidf_matrix)
movie_similarity
```

```python
#reverse mapping 
index = pd.Series(movie_credit.index,index=movie_credit['title_x']).drop_duplicates()
index.head()
```

```python
#function for recommendation based on title
def recommend(title,movie_similarity=movie_similarity):
    #getting index of title
    idx = index[title]
    #similarity score
    similarity_score=list(enumerate(movie_similarity[idx]))
    #sorting based on similarity score
    similarity_score=sorted(similarity_score,key=lambda x:x[1],reverse=True)
    #getting 10 similar movies
    score=similarity_score[1:11]
    #getting indices
    movie_idx= [i[0] for i in similarity_score[1:11]]
    return movie_credit['title_x'].iloc[movie_idx]


```
